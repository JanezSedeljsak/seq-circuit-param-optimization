## Finding optimal parameters with the Genetic algorithm

```python
best_params = GeneticAlgorithm(FreqEvaluator).optimize_parameters(40, 15)
evaluation = FreqEvaluator(**best_params)

print(evaluation.get_params())
print(evaluation.evaluate())
evaluation.simulate()
```

```python
best_params = GeneticAlgorithm(ClockEvaluator).optimize_parameters(40, 15)
evaluation = ClockEvaluator(**best_params)

print(evaluation.get_params())
print(evaluation.evaluate())
evaluation.simulate()
```

## Finding optimal parameters with the Ant colony optimization algorithm

```python
best_params = AntColonyOptimization(FreqEvaluator, num_ants=10, max_iterations=30).optimize_parameters()
evaluation = FreqEvaluator(**best_params)

print(evaluation.get_params())
print(evaluation.evaluate())
evaluation.simulate()
```

```python
best_params = AntColonyOptimization(ClockEvaluator, num_ants=10, max_iterations=30).optimize_parameters()
evaluation = ClockEvaluator(**best_params)

print(evaluation.get_params())
print(evaluation.evaluate())
evaluation.simulate()
```

## Joined evaluation with ant colony optimization

```python
best_params = AntColonyOptimization(
    JoinedEvaluator,
    joined=[ClockEvaluator, FreqEvaluator],
    num_ants=20,
    max_iterations=100
).optimize_parameters()

evaluation = EvaluationBase(**best_params)
# here we can't print the score because it's caclulated based on 2 evaluators
print(evaluation.get_params())
evaluation.simulate()
```
